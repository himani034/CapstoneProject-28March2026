```python
from groq import Groq
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
client = Groq(api_key=os.getenv("GROQ_API_KEY"))
```

```python
def llm_call(prompt):
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",  # updated to a supported model
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content
```

```python
# Agent 1: Classifier Agent
def classifier_agent(query):
    prompt = f"""
    Classify the user query into ONLY one category:
    Billing, Technical, or General.

    Query: {query}

    Answer in one word only.
    """
    return llm_call(prompt).strip().lower()
```

```python

# Agent 2: Billing Agent
def billing_agent(query):
    prompt = f"""
    You are a billing support agent.
    Handle this query professionally:
    {query}
    """
    return llm_call(prompt)
```

```python
# Agent 3: Technical Agent
def technical_agent(query):
    prompt = f"""
    You are a technical support agent.
    Solve the issue:
    {query}
    """
    return llm_call(prompt)
```

```python
# Agent 4: Response Agent
def response_agent(category, query):
    if "billing" in category:
        return billing_agent(query)
    elif "technical" in category:
        return technical_agent(query)
    else:
        return llm_call(f"Answer this general query professionally: {query}")
```

```python
def support_system(query):
    print("\nUser Query:", query)
    
    # Step 1: Classification
    category = classifier_agent(query)
    print("Classified as:", category)
    
    # Step 2: Delegation
    response = response_agent(category, query)
    
    # Step 3: Final Output
    print("Final Response:\n", response)
    print("-" * 50)

queries = [
    "I was charged twice for my subscription",
    "The app crashes when I open it",
    "What are your pricing plans?"
]

for q in queries:
    support_system(q)
```
